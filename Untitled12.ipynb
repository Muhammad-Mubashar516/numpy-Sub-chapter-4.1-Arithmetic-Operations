```python
import numpy as np

# Demo arrays
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Array a:", a)
print("Array b:", b)


# ============================================================
# Topic 1: Addition (np.add)
# ============================================================
print("\n=== Topic 1: Addition (np.add) ===")
print(np.add(a, b))     # Same as a + b


# ============================================================
# Topic 2: Subtraction (np.subtract)
# ============================================================
print("\n=== Topic 2: Subtraction (np.subtract) ===")
print(np.subtract(a, b))   # Same as a - b


# ============================================================
# Topic 3: Multiplication (np.multiply)
# ============================================================
print("\n=== Topic 3: Multiplication (np.multiply) ===")
print(np.multiply(a, b))   # Same as a * b


# ============================================================
# Topic 4: Division (np.divide)
# ============================================================
print("\n=== Topic 4: Division (np.divide) ===")
print(np.divide(a, b))     # Float result


# ============================================================
# Topic 5: Floor Division (np.floor_divide)
# ============================================================
print("\n=== Topic 5: Floor Division (np.floor_divide) ===")
print(np.floor_divide(a, b))


# ============================================================
# Topic 6: Modulus (np.mod)
# ============================================================
print("\n=== Topic 6: Modulus (np.mod) ===")
print(np.mod(a, b))        # Same as a % b


# ============================================================
# Topic 7: Power (np.power)
# ============================================================
print("\n=== Topic 7: Power (np.power) ===")
print(np.power(a, b))      # a^b element-wise


# ============================================================
# Topic 8: Element-wise Operations
# ============================================================
print("\n=== Topic 8: Element-wise Operations ===")
print("a + 5 =", a + 5)    # scalar with array
print("a * 2 =", a * 2)
print("a ** 2 =", a ** 2)


# ============================================================
# Topic 9: Broadcasting Basics
# ============================================================
print("\n=== Topic 9: Broadcasting Basics ===")
c = np.array([[1, 2, 3],
              [4, 5, 6]])
print("Array c:\n", c)
print("c + 10:\n", c + 10)   # scalar broadcast


# ============================================================
# Topic 10: Broadcasting Rules
# ============================================================
print("\n=== Topic 10: Broadcasting Rules ===")
d = np.array([1, 2, 3])     # shape (3,)
print("c shape:", c.shape)
print("d shape:", d.shape)
print("c + d:\n", c + d)    # Works due to broadcasting


# ============================================================
# Topic 11: Broadcasting Examples
# ============================================================
print("\n=== Topic 11: Broadcasting Examples ===")
e = np.array([[1], [2], [3]])   # shape (3,1)
f = np.array([10, 20, 30])      # shape (3,)
print("e:\n", e)
print("f:", f)
print("e + f:\n", e + f)        # broadcast row + col
```
